```python
from collections import OrderedDict
from functools import partial
from time import time

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter

from sklearn import manifold, datasets
```

```python
n_neighbors, n_components = 1, 2
n_components = 2
color = np.arange(-5, 5, 1)
```

```python
n_neighbors, n_components = 1, 2
methods = OrderedDict()
methods['Isomap'] = manifold.Isomap(n_neighbors, 
                                    n_components)
methods['MDS'] = manifold.MDS(n_components, 
                              max_iter=100, 
                              n_init=1)
methods['t-SNE'] = manifold.TSNE(n_components=n_components, 
                                 init='pca',
                                 random_state=0)
```

```python
name_list = ['tshirt', 'trouser', 'pullover', 'dress', 'coat',
             'sandal', 'shirt', 'sneaker', 'bag', 'boot']
```

```python
X = oc_gram_sys
fig = plt.figure(figsize=(15, 4))
# fig.suptitle('Manifold Embeddings for Distance Matrix (Latent Space from One-Class Model)')

for i, (label, method) in enumerate(methods.items()):
    t0 = time()
    Y = method.fit_transform(X)
    t1 = time()
    print("%s: %.2g sec" % (label, t1 - t0))
    ax = fig.add_subplot(1, 3, i + 1)
    scatter_ = ax.scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
    
    ax.set_title(label)
#     ax.xaxis.set_major_formatter(NullFormatter())
#     ax.yaxis.set_major_formatter(NullFormatter())
    ax.axis('tight')
    
fig.legend(handles=scatter_.legend_elements()[0], 
           labels=name_list,
           loc='center right',
           borderaxespad=0.1,
           title='classes')
```
